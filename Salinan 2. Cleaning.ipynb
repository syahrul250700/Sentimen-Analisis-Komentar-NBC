### 2. Menghapus Karakter User

```python
#remove data user
def remove_pattern(facebook, pattern):
    r = re.findall(pattern, facebook)
    for i in r:
        facebook = re.sub(i, '', facebook)
    return facebook    
df['remove_user'] = np.vectorize(remove_pattern)(df['Komentar'], "@[\w]*")
```

### 3. Mengapus Karakter Angka dan Hastag

```python
  def remove(facebook):
    #remove angka
    facebook = re.sub('[0-9]+', '', facebook)
    
    # remove stock market tickers like $GE
    facebook = re.sub(r'\$\w*', '', facebook)
 
    # remove hashtags
    # only removing the hash # sign from the word
    facebook = re.sub(r'#', '', facebook)
    return facebook
df['remove_angkadanhastag'] = df['remove_user'].apply(lambda x: remove(x))
```

```python
#import stopword
from nltk.corpus import stopwords 
stopwords_indonesia = stopwords.words('indonesian')

#import Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()

#tokenize
from nltk.tokenize import TweetTokenizer

# Happy Emoticons
emoticons_happy = set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'
    ])
 
# Sad Emoticons
emoticons_sad = set([
    ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('
    ])
 
# all emoticons (happy + sad)
emoticons = emoticons_happy.union(emoticons_sad)

def clean_facebooks(facebook):
    # remove stock market tickers like $GE
    facebook = re.sub(r'\$\w*', '', facebook)
 
    # remove old style facebook text "RT"
    facebook = re.sub(r'^RT[\s]+', '', facebook)
 
    # remove hyperlinks
    facebook = re.sub(r'https?:\/\/.*[\r\n]*', '', facebook)
    
    # remove hashtags
    # only removing the hash # sign from the word
    facebook = re.sub(r'#', '', facebook)
    
    #remove coma
    facebook = re.sub(r',','',facebook)
    
    #remove angka
    facebook = re.sub('[0-9]+', '', facebook)
 
    # tokenize facebooks
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    facebook_tokens = tokenizer.tokenize(facebook)
 
    facebooks_clean = []    
    for word in facebook_tokens:
        if (word not in stopwords_indonesia and # remove stopwords
              word not in emoticons and # remove emoticons
                word not in string.punctuation): # remove punctuation
            #facebooks_clean.append(word)
            stem_word = stemmer.stem(word) # stemming word
            facebooks_clean.append(stem_word)
 
    return facebooks_clean
   
```

```python
 df['facebooks_clean'] = df['remove_angkadanhastag'].apply(lambda x: clean_facebooks(x))
```

```python
#Stop Removal

stopword = nltk.corpus.stopwords.words('indonesian')

def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

df['Stop_Removal'] = df['facebooks_clean'].apply(lambda x: remove_stopwords(x))
df
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}
for document in df['Stop_Removal']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)

# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['komen_stemmed'] = df['Stop_Removal'].apply(get_stemmed_term)
```
