```python
tel['Facebook'] = [str(tw) for tw in tel['Facebook']]
#vector = CountVectorizer()
vector_tf = TfidfVectorizer(max_df=0.9, min_df=0.02)
#matrix_kata =  vector.fit_transform(examp.stemming)
matrix_kata_tf = vector_tf.fit_transform(tel.Facebook)
matrix_kata_tf.shape
```

```python
data = pd.DataFrame(matrix_kata_tf.toarray(), columns=vector_tf.get_feature_names())
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
X_train, X_test, y_train, y_test = train_test_split(data, tel.sentimen, test_size=0.3, random_state=42)

print(y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
y_test_pred=clf.predict(X_test)
confusion_matrix(y_test,y_test_pred)
skor_prediksi = clf.predict_proba(X_test)
hasil_prediksi = clf.predict(X_test)
print(classification_report(y_test,y_test_pred))
clf.score(X_test,y_test)
print(confusion_matrix(y_test,y_test_pred))
```
