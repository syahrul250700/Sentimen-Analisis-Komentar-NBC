```python
import pandas as pd
import numpy as np
import nltk 
import string
import re
import os
```

```python
#pemanggilan data mentah
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/dataset_preprocessed.xlsx')
```

```python
pos_sent = open("/content/drive/MyDrive/Colab Notebooks/kamus sentimen/positive.txt").read()
positive_words=pos_sent.split('\n')
```

```python
neg_sent = open("/content/drive/MyDrive/Colab Notebooks/kamus sentimen/negative.txt").read()
negative_words=neg_sent.split('\n')
```

```python
data['Facebook'] = [str(fb) for fb in data['Facebook']]
```

```python
data['count'] = data['Facebook'].apply(lambda Facebook: sum(0 + ((word in positive_words) and 1) + ((word in negative_words) and -1) for word in Facebook.split()))
```

```python
data['sentimen'] = 'netral'
data['sentimen'][data['count']>0] = 'positif'
data['sentimen'][data['count']<0] = 'negatif'
data['sentimen'][data['count']==0] = 'netral'
```
